```python
import nibabel as nib
import numpy as np
import os
```

# CIMT

```python
experiment = 'LLstim'
if experiment == 'rs':
    pass
elif experiment == 'LLstim':
    experiment_path = '/Users/fei/Downloads/LLstim'
    raw_nii_dir_name = '03_0_stats_00_rsfmri_resid_masked_STANDARDSPACE' # just the name of the directory, not the full path
    raw_nii_path = os.path.join(experiment_path, raw_nii_dir_name)
    scaled_nii_path = os.path.join(experiment_path, raw_nii_dir_name + '_scaled')
    os.mkdir(scaled_nii_path) # use .mkdir() instead of .makedirs() so an error is raised if directory already exists
    for f in sorted(os.listdir(raw_nii_path)):
        if f != '.DS_Store':
            nii = nib.load(os.path.join(raw_nii_path, f))
            nii_data = nii.get_fdata()
            nii_data = np.nan_to_num(100 + (nii_data - np.mean(nii_data, axis=3, keepdims=True)) / np.std(nii_data, axis=3, keepdims=True))
            scaled_nii = nib.Nifti1Image(nii_data, nii.affine, header=nii.header)
            f_split = f.split('.') # for naming purposes
            nib.save(scaled_nii, os.path.join(scaled_nii_path, f'{f_split[0]}_scaled.{f_split[1]}.{f_split[2]}'))
```
